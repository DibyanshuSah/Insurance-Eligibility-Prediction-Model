```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("C:/Users/dibya/OneDrive/Desktop/Files_Jupyter Notebook/insurance_data_5000.csv")
df.head()
```

```python
a=df["age"]
b=df["bought_insurance"]

plt.xlabel("age")
plt.ylabel("bought_insurance")
plt.scatter(a,b,marker="+",color="g")
plt.title("Insurance approval graph")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[["age"]],b,train_size=0.95)
```

```python
x_test.head()
```

```python
x_train.head(3)
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
log_reg.predict([[65]])
```

```python
log_reg.predict(x_test)
```

```python
log_reg.predict(x_test.head())
```

```python
print("Accuracy of the model is : ",log_reg.score(x_test,y_test))
```
